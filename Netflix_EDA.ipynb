### 3. Who are the most popular directors and actors on Netflix?

```python
from collections import Counter
import math
```

```python
x = data['country'].value_counts().head(6)
x.index
```

```python
colors = ["orangered", "mediumseagreen", "darkturquoise", "mediumpurple", "deeppink", "indianred"]
countries = ['United States', 'India', 'United Kingdom', 'Canada', 'Spain', 'Egypt']
```

```python
with plt.xkcd():
    plt.figure(figsize = (20, 8))
    x = 1
    for c in countries:
        country_data = data[data['country'] == c]
        cat = ", ".join(country_data["director"].fillna("")).split(", ")
        d_list = Counter(cat).most_common(6)
        category1 = [i[0] for i in d_list][::-1]
        value1 = [i[1] for i in d_list][::-1]
        if max(value1) < 10:
            int_val = range(0, math.ceil(max(value1))+1)
        else:
            int_val = range(0, math.ceil(max(value1))+1, 2)
        plt.subplot(2, 3, x)
        plt.barh(category1, value1, color = colors[x-1])
        plt.xticks(int_val)
        plt.title(c)
        x += 1
    plt.suptitle('Popular directors with the most content')
    plt.tight_layout()
    plt.show()
```

```python
data['cast']
```

```python
with plt.xkcd():
    x = 1
    plt.figure(figsize = (20,8))
    for i,c in enumerate(countries, start = 1):
        small = data[data['country'].fillna('').str.lower().str.contains(c.lower())]
        cast = ', '.join(small['cast'].fillna('')).split(',')
        a_list = Counter(cast).most_common(10)
        cat1 = [i[0] for i in a_list][::-1]
        val1 = [i[1] for i in a_list][::-1]
        if max(val1) < 10:
            val1_int = range(0, math.ceil(max(val1)) + 1)
        elif max(val1) >= 10 and max(val1) <= 20:
            val1_int = range(0, math.ceil(max(val1)) + 1, 2)
        else:
            val1_int = range(0, math.ceil(max(val1)) + 1, 5)
        plt.subplot(2, 3, x)
        plt.barh(cat1, val1, color = colors[x-1])
        plt.xticks(val1_int)
        plt.title(c)
        x += 1
    plt.suptitle('Popular actors with the most content')
    plt.tight_layout()
    plt.show()
```

### 5. What are the top genres per country? 

```python
colors = ["orangered", "mediumseagreen", "darkturquoise", "mediumpurple", "deeppink", "indianred"]
countries = ['United States', 'India', 'United Kingdom', 'Canada', 'Spain', 'Egypt']
col = 'listed_in'
```

```python
with plt.xkcd():
    x = 1
    plt.figure(figsize = (20,8))
    for c in countries:
        genre_data = data[data['country'].str.contains(c)]
        genre_data = genre_data.dropna(subset = col)
        genre = ', '.join(genre_data[col]).split(', ')
        gen_list = Counter(genre).most_common(4)
        gen_list = [i for i in gen_list if '' != i[0]]
        cat2 = [i[0] for i in gen_list][::-1]
        val2 = [i[1] for i in gen_list][::-1]
        if max(val2) < 100:
            val2_int = range(0, math.ceil(max(val2)), 20)
        elif max(val2) > 100:
            val2_int = range(0, math.ceil(max(val2)) + 50, 100)
        else:
            val2_int = range(0, math.ceil(max(val2)) + 25, 200)
        plt.subplot(3, 2, x)
        plt.barh(cat2, val2, color = colors[x-1])
        plt.xticks(val2_int)
        plt.title(c)
        x += 1
    plt.suptitle('Top Genres per Country')
    plt.tight_layout()
    plt.show()
```
